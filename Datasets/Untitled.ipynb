```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("lag_valid_dataset.csv")
predicted = pd.read_csv("predicted_dataset.csv")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['open_price_t5'], label='Open Price T5', color='blue', marker='o')
plt.plot(predicted['time'], predicted['open_price_t5'], label='Open Price T5', color='red', marker='o')
# Adding labels and title
plt.xlabel('Time')
plt.ylabel('Open Price T5')
plt.title('Time vs Open Price T5')
plt.grid(True)
plt.legend()

# Rotate time labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
